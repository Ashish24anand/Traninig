```python
import pandas as pd
import selenium as se
from selenium import webdriver
import time
from selenium.webdriver.support.select import Select
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(3)

url = "https://www.amazon.in"
driver.get(url)
time.sleep(2)
# user input for search
search=str(input("enter item to search"))


#locating the search bar
search_bar = driver.find_element_by_id("twotabsearchtextbox")    # Locating searc_bar by id
search_bar.clear()                                               # clearing search_bar
search_bar.send_keys(search)                                   # sending user input to search bar
search_button = driver.find_element_by_xpath('//span[@id="nav-search-submit-text"]')       # Locating search_button by xpath
search_button.click()    
```

```python
def scraping_page():
    #creating the empty list
    Brand_Name=[]
    Name_of_the_Product=[]
    product_details=[]
    Rating=[]
    No_of_Ratings=[]
    Price=[]
    Return_Exchange=[]
    Expected_Delivery=[]
    Availability=[]
    Product_URL=[]
    time.sleep(2)
        
    df=pd.DataFrame()

    #Scrapping product_url
    url=driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
    for i in url:
        Product_URL.append(i.get_attribute('href'))
        
        
    
            
        

    #Scrapping product details
    pro_detail=driver.find_elements_by_xpath("//span[@class='a-size-base-plus a-color-base a-text-normal']")
    for i in pro_detail:
        product_details.append(i.text)
        

    for i in Product_URL:
        driver.get(i)
        time.sleep(2)
        #scrapping Price
        try:
            prices=driver.find_element_by_id("priceblock_ourprice")
            Price.append(prices.text)
        
        except NoSuchElementException:
            Price.append("Not mentioned")
        
        # scrapping rating  
        try:                  #exception handling for nosuchelementexception                                                    #click the rating link found
            rating=driver.find_element_by_xpath("//span[@class='a-size-base a-nowrap']//span")#locating the rating
            Rating.append(rating.text)#appending the ratings in Ratings list
        
        except NoSuchElementException:
            Rating.append("NO rating")#appending the No rating if no rating is there
    
        
    
    #Scrapping product details
        try:
            pro_detail=driver.find_element_by_id("feature-bullets")
            product_details.append(pro_detail.text)
        except NoSuchElementException:
            product_details.append("NO details")
        
        
    #scraping no of rating
        try:                  #exception handling for nosuchelementexception                                                    #click the rating link found
            n_rating=driver.find_element_by_id("acrCustomerReviewText")#locating the Number of rating
            No_of_Ratings.append(n_rating.text)#appending the ratings in No of Ratings list
        
        except NoSuchElementException:
            No_of_Ratings.append("NO rating")
        
        #scraping brand name
        try:                  #exception handling for nosuchelementexception                                                    #click the rating link found
            brand_name=driver.find_element_by_xpath("//span[@class='a-size-base']")#locating the brand name
            Brand_Name.append(brand_name.text)#appending the brand name 
        
        except NoSuchElementException:
            Brand_Name.append("NO rating")
        
    
        
    #scraping product title
        try:                  #exception handling for nosuchelementexception                                                    #click the rating link found
            pro_name=driver.find_element_by_id("productTitle")#locating the brand name
            Name_of_the_Product.append(pro_name.text.split('|')[0])#appending the brand name 
        
        except NoSuchElementException:
            Name_of_the_Product.append("NO rating")
        
    
        
    #scraping return 
        try:                  #exception handling for nosuchelementexception                                                    #click the rating link found
            pro_exchange=driver.find_element_by_xpath("//a[@Class='a-size-small a-link-normal a-text-normal']")#locating the Return/Exchange 
            Return_Exchange.append(pro_exchange.text)#appending the Return/Exchange  
        
        except NoSuchElementException:
            Return_Exchange.append("NO rating")
        
    
    #scrapping Expected delivery    
        try:                  #exception handling for nosuchelementexception                                                    #click the rating link found
            ex_dil=driver.find_element_by_id("ddmDeliveryMessage")#locating the Return/Exchange 
            Expected_Delivery.append(ex_dil.text)#appending the Return/Exchange  
        
        except NoSuchElementException:
            Expected_Delivery.append("NO rating")
        
        
    #scrapping Avilability
        try:                  #exception handling for nosuchelementexception                                                    #click the rating link found
            ava=driver.find_element_by_id("availability")#locating the Return/Exchange 
            Availability.append(ava.text)#appending the Return/Exchange  
        
        except NoSuchElementException:
            Availability.append("NO rating")
        
        

        
        
    
        


#creating a dataframe
    df=pd.DataFrame({'Brand Name':Brand_Name,
              'Name of the Product':Name_of_the_Product,
              'product details':product_details[:len(Product_URL)],
              'Rating':Rating,
              'No of Ratings':No_of_Ratings,
              'Price':Price,
              'Return Exchange':Return_Exchange,
              'Expected Delivery':Expected_Delivery,
              'Availability':Availability,
              'Product URL':Product_URL})


#printing dataframe
    return(df)
```

Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Display Resolution”, “Processor”, “Processor Cores”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV.

```python
brand_name=[]
phone_name=[]
colour=[]
ram=[]
rom=[]
primary_camera=[]
secondary_camera=[]
display_size=[]
display_resolution=[]
processor=[]
processor_cores=[]
battery_capacity=[]
price=[]
product_URL=[]
driver=webdriver.Chrome("chromedriver.exe")
time.sleep(2)
driver.get('https://www.flipkart.com/')
time.sleep(2)
close=driver.find_element_by_xpath("//button[@Class='_2KpZ6l _2doB4z']")
close.click()
search_bar=driver.find_element_by_xpath("//input[@Class='_3704LK']")
search_bar.clear()
search_bar.send_keys('mobile')
search=driver.find_element_by_xpath("//button[@Class='L0Z3Pu']")
search.click()
time.sleep(2)
img=driver.find_elements_by_xpath("//a[@Class='s1Q9rs']")
for i in img:
    product_URL.append(i.get_attribute('href'))
    
for i in product_URL:
    driver.get(i)
    time.sleep(2)
    driver.find_element_by_xpath("//button[@Class='_2KpZ6l _1FH0tX']").click()# finding and clicking read more
    try:
        pro_name=driver.find_element_by_xpath("//a[@Class='_2whKao']")#locating the brand name
        brand_name.append(pro_name.text)#appending the brand name 
        
    except NoSuchElementException:
        Name_of_the_Product.append("-")
        
    try:
        ph_name=driver.find_element_by_xpath("//span[@Class='B_NuCI']")#locating the phone name
        try:
            phone_name.append(ph_name.text.split('(')[0])#appending the phone name
        except:
            phone_name.append('-')
                
        try:
            colour.append(ph_name.text.split('(')[1].split(',')[0])#appending the color
        except:
            colour.append('-')
                
        try:
            rom.append(ph_name.text.split('(')[1].split(',')[1])#appending the rom
        except:
            rom.append('-')
                
        try:
            ram.append(ph_name.text.split('(')[-1].split(')')[0])#appending the ram
        except:
            ram.append('-')
        
    except NoSuchElementException:
        phone_name.append("-")
        colour.append("-")
        rom.append("-")
        ram.append("-")
            
    try:
        cam=driver.find_elements_by_xpath("//li[@Class='_21lJbe']")
        di=0
        ci=0
        for i in cam:
            try:
                if i.text.__contains__('Pixels') and di==0:
                    display_resolution.append(i.text)
                    di=1
            except:
                display_resolution.append('-')
                di=1
            try:
                if i.text.__contains__('Core') and ci==0:
                    processor_cores.append(i.text)
                    ci=1
            except:
                processor_cores.append('-')
                ci=1
        if di==0:
            display_resolution.append('-')
        if ci==0:
            processor_cores.append('-')
        
    except NoSuchElementException:
        display_resolution.append('-')
        processor_cores.append('-')
    
        
    try:
        pro_price=driver.find_element_by_xpath("//div[@Class='_30jeq3 _16Jk6d']")#locating the price
        price.append(pro_price.text)#appending the price 
        
    except NoSuchElementException:
        price.append("-")
            
    try:
        pro_pro=driver.find_elements_by_xpath("//li[@Class='_21Ahn-']")#locating the details
        try:
            if pro_pro[4].text.__contains__('Processor'):
                processor.append(pro_pro[4].text)#appending the price 
        except:
            processor.append("-")
        
        
        try:
            if pro_pro[2].text.split('|')[0].__contains__('MP'):
                primary_camera.append(pro_pro[2].text.split('|')[0])
            else:
                primary_camera.append('-')
        except:
            primary_camera.append('-')
        try:
            if pro_pro[2].text.split('|')[1].__contains__('Front Camera'):
                secondary_camera.append(pro_pro[2].text.split('|')[1])
            else:
                secondary_camera.append('-')
        except:
            secondary_camera.append('-')
        
        try:
            if pro_pro[1].text.__contains__('cm'):
                display_size.append(pro_pro[1].text)
            else:
                display_size.append('-')
        except:
            display_size.append('-')
            
        try:
            if pro_pro[3].text.__contains__('mAh'):
                battery_capacity.append(pro_pro[3].text)
            else:
                battery_capacity.append('-')
        except:
            battery_capacity.append('-')
            
    
    except NoSuchElementException:
        processor.append("-")
        battery_capacity.append('-')
        display_size.append('-')
        secondary_camera.append('-')
        primary_camera.append('-')
        processor.append("-")
        
```

```python
df=pd.DataFrame({   'brand name':brand_name,
                    'phone name':phone_name,
                    'colour':colour,
                    'ram':ram,
                    'rom':rom,
                    'primary_camera':primary_camera,
                    'secondary_camera':secondary_camera,
                    'display size':display_size,
                    'display resolution':display_resolution,
                    'processor':processor,
                    'processor_core':processor_cores,
                    'battery capacity':battery_capacity,
                    'price':price,
                    'product URL':product_URL})
df
```
