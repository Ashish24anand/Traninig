Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(3)

url = "https://github.com/"
driver.get(url)
time.sleep(2)
li=driver.find_elements_by_xpath("//summary[@Class='HeaderMenu-summary HeaderMenu-link px-0 py-3 border-0 no-wrap d-block d-lg-inline-block']")
for i in li:
    if i.text == 'Explore':
        i.click()
li1=i.find_elements_by_xpath("//a[@Class='py-2 lh-condensed-ultra d-block Link--primary no-underline h5 Bump-link--hover']")

driver.get(li1[1].get_attribute('href'))
time.sleep(2)
li=driver.find_elements_by_xpath('//a[@Class="js-selected-navigation-item d-inline-block py-2 py-md-3 mr-3 mr-md-4 no-underline subnav-link"]')
for i in li:
    if i.text=='Trending':
        u=i.get_attribute('href')
        driver.get(u)
        break
time.sleep(2)
```

```python
Repository_title=[]
Repository_description=[]
Contributors_count=[]
Language_used=[]
li1=driver.find_elements_by_xpath('//article[@Class="Box-row"]')
for j in li1:
    try:
        li=j.find_element_by_xpath('.//span[@Class="text-normal"]')
        Repository_title.append(li.text.split('/')[0])
    except NoSuchElementException:
        Repository_title.append('-')
    try:
        li=j.find_element_by_xpath('.//p[@Class="col-9 color-text-secondary my-1 pr-4"]')
        Repository_description.append(li.text)
    except NoSuchElementException:
        Repository_description.append('-')
    
    try:
        li=j.find_elements_by_xpath('.//a[@Class="Link--muted d-inline-block mr-3"]')
        for i in range(len(li)):
            if i % 2 != 0:
                Contributors_count.append(li[i].text)
    except NoSuchElementException:
        Contributors_count.append('-')
    try:
        li=j.find_element_by_xpath('.//span[@itemprop="programmingLanguage"]')
        Language_used.append(li.text)
    except NoSuchElementException:
        Language_used.append('-') 
        
driver.close()
    
df=pd.DataFrame({'Repository_title':Repository_title,'Repository_description':Repository_description,'Contributors_count':Contributors_count,'Language_used':Language_used})
df
```

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(3)

url = "https://www.billboard.com/"
driver.get(url)
time.sleep(2)
li=driver.find_elements_by_xpath('//a[@Class="header__subnav__link text--uppercase font--semi-bold"]')
for i in li:
    if i.text == 'HOT 100':
        u=i.get_attribute('href')
        driver.get(u)
        break
```

```python
Song_name=[]
Artist_name=[]
Last_week_rank=[]
Peak_rank=[]
Weeks_on_board=[]
li1=driver.find_elements_by_xpath('//button[@Class="chart-element__wrapper display--flex flex--grow sort--this-week"]')
for j in li1:
    try:
        li=j.find_element_by_xpath('.//span[@Class="chart-element__information__song text--truncate color--primary"]')
        Song_name.append(li.text)
    except NoSuchElementException:
        Song_name.append('-')
    try:
        li=j.find_element_by_xpath('.//span[@Class="chart-element__information__artist text--truncate color--secondary"]')
        Artist_name.append(li.text)
    except NoSuchElementException:
        Artist_name.append('-')
    
    try:
        li=j.find_element_by_xpath('.//div[@Class="chart-element__meta text--center color--secondary text--last"]')
        Last_week_rank.append(li.text)
    except NoSuchElementException:
        Last_week_rank.append('-')
    try:
        li=j.find_element_by_xpath('.//div[@Class="chart-element__meta text--center color--secondary text--peak"]')
        Peak_rank.append(li.text)
    except NoSuchElementException:
        Peak_rank.append('-') 
        
    try:
        li=j.find_element_by_xpath('.//div[@Class="chart-element__meta text--center color--secondary text--week"]')
        Weeks_on_board.append(li.text)
    except NoSuchElementException:
        Weeks_on_board.append('-') 
    
driver.close()

df=pd.DataFrame({'Song_name':Song_name,'Artist_name':Artist_name,'Last_week_rank':Last_week_rank,'Peak_rank':Peak_rank,'Weeks_on_board':Weeks_on_board})
df
```

Details of Datasets from UCI machine learning repositories.
Url = https://archive.ics.uci.edu/
You have to find the following details:
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute
G) Year
Note: - from the home page you have to go to the Show All Dataset page through code

```python
driver=webdriver.Chrome("chromedriver.exe") 
time.sleep(4)

url = "https://archive.ics.uci.edu/"
driver.get(url)
time.sleep(2)
li=driver.find_elements_by_xpath('//span[@Class="normal"]//a')
for i in li:
    if i.text == 'view all data sets':
        u=i.get_attribute('href')
        driver.get(u)
        break
time.sleep(2)
```

```python
Dataset_name=[]
Data_type=[]
Task=[]
Attribute_type=[]
No_of_instances=[]
No_of_attribute=[]
Year=[]

li1=driver.find_elements_by_xpath('//table[@border="1"]//tr')
o=0
for i in li1:
    #li=i.find_element_by_xpath('.//p[@Class="normal"]//a')
    #Dataset_name.append(li.text)
    li=i.find_elements_by_xpath('.//p[@Class="normal"]')
    for j in range(len(li)):
        if j == 0:
            Dataset_name.append(li[j].text)
            o=o+1
        elif j == 1:
            Data_type.append(li[j].text)
            o=o+1
        elif j == 2:
            Task.append(li[j].text)
            o=o+1
        elif j == 3:
            Attribute_type.append(li[j].text)
            o=o+1
        elif j == 4:
            No_of_instances.append(li[j].text)
            o=o+1
        elif j == 5:
            No_of_attribute.append(li[j].text)
            o=o+1
        else:
            o=o+1
dn=[]
for i in range(len(Dataset_name)):
    if i%2 == 0:
        dn.append(Dataset_name[i])
driver.close()
df=pd.DataFrame({'Dataset_name':dn,'Data_type':Data_type,'Task':Task,'Attribute_type':Attribute_type,'No_of_instances':No_of_instances,'No_of_attribute':No_of_attribute})
df
```
