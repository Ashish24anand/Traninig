```python
job_dis=pd.DataFrame()
j=0
for lnk in url:
    driver.get(lnk)
    dis=driver.find_elements_by_xpath("//section[@class='job-desc']")
    k=[]
    for i in dis:
        k.append(i.text)
    
    job_dis=pd.concat([job_dis,pd.DataFrame(k)],axis=0,ignore_index=True)
    
    dis=driver.find_elements_by_xpath("//div[@class='nConfig-jd-heading-container']")
    k=[]
    for i in dis:
        k.append(i.text)
    
    job_dis=pd.concat([job_dis,pd.DataFrame(k)],axis=0,ignore_index=True)
    if j>9:
        break
    j=j+1

    
```

```python
job_dis=job_dis.head(10)
```

```python
d_job=pd.DataFrame([job_title,job_loc,com_name]).T.head(10)
```

## Q6 Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:

```python
driver=webdriver.Chrome("chromedriver.exe")

url= 'https://www.flipkart.com/'
driver.get(url)

close=driver.find_element_by_xpath("//button[@class='_2KpZ6l _2doB4z']")
close.click()

search_bar=driver.find_element_by_xpath("//input[@title='Search for products, brands and more']")
search_bar.send_keys('sunglasses')

search=driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search.click()

time.sleep(5)
```

```python
def sunglass():
    brand=driver.find_elements_by_xpath("//div[@class='_2WkVRV']")
    k=[]
    for i in brand:
        k.append(i.text)
    
    brand_name=k

    dis=driver.find_elements_by_xpath("//a[@class='IRpwTa']")

    k=[]
    for i in dis:
        k.append(i.text)
    
    pro_discrip=k

    price=driver.find_elements_by_xpath("//div[@class='_30jeq3']")

    k=[]
    for i in price:
        k.append(i.text)
    
    pro_price=k

    discount=driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']")

    k=[]
    for i in discount:
        k.append(i.text.split('o')[0])
    
    pro_dis=k

    return(pd.DataFrame([brand_name,pro_discrip,pro_price,pro_dis]).T)
```

```python
demo=pd.DataFrame()
i=0
demo1=sunglass()
demo=pd.concat([demo,demo1],axis=0,ignore_index=True)
next=driver.find_element_by_xpath("//a[@class='_1LKTO3']")
next.click()
i=demo.index.max()
time.sleep(5)
while i<100:
    demo1=sunglass()
    demo=pd.concat([demo,demo1],axis=0,ignore_index=True)
    next=driver.find_elements_by_xpath("//a[@class='_1LKTO3']")
    next[1].click()
    i=demo.index.max()
    time.sleep(5)
    
```

```python
demo.columns=['Brand','Product Description','Price','Discount %']
```

```python
demo.iloc[:100]
```

```python
def laptop():
    title=driver.find_elements_by_xpath("//span[@class='a-size-medium a-color-base a-text-normal']")
    k=[]
    for i in title:
        k.append(i.text)
    
    title=k[:5]

    rating=driver.find_elements_by_xpath("//span[@class='a-icon-alt']")

    k=[]
    for i in rating:
        if i.text=='':
            k.append('-')
        else:
            k.append(i.text)
            print(i.text)
        
    
    rating=k[:5]

    price=driver.find_elements_by_xpath("//span[@class='a-price-whole']")

    k=[]
    for i in price:
        k.append(i.text)
    
    pro_price=k[:5]
    return(pd.DataFrame([title,rating,pro_price]).T)
```
